```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
import os
```

```python
df = pd.read_csv('/home/ubuntu/Github/LABS/Data/cleanedcitydata.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
# citylist = df['city', 'City_Name', 'State']
```

```python
df['Population_Density'].describe()
```

```python
df['popden_score'] = abs(df['Population_Density'] - 1304)/1304
df['popden_score'].describe()
```

```python
den = sns.scatterplot(x='Population_Density', y='popden_score', data=df)
```

```python
den1 = sns.violinplot(x='Population_Density', data=df)
```

```python
df['popden_norm'] = st.norm.cdf((df['popden_score'].mean() - df['popden_score'])/df['popden_score'].std()) * 100
df['popden_norm'] = df['popden_norm'].round(1)
df['popden_norm'].describe()

# a, b = 35, 0
# x, y = df['popden_raw'].min(), df['popden_raw'].max()
# df['popden_norm'] = (df['popden_raw'] - x) / (y -x) * (b -a) + a
# df['popden_norm'].describe()
```

```python
housevalue = df['Median_House_Value']
tax = df['Property_taxes']
med_income = df['Median_Income']

downpct = .115
loanamt = housevalue * (1 - downpct)
monthlyint = 0.003325
numpmts = 360

mrtgpmts = loanamt * (monthlyint / (1 - ((1 + monthlyint)**(-numpmts)))) *12

ins = housevalue * .0035
pmi = loanamt * .01

housingpmts = mrtgpmts + tax + ins + pmi

df['HDTI'] = housingpmts / med_income * 100

df['HDTIscore'] = 28 - df['HDTI']
```

```python
df['HDTIscore'].describe()
```

```python
df['HDTIscorenorm'] = st.norm.cdf((df['HDTIscore'] - df['HDTIscore'].mean())/df['HDTIscore'].std()) * 100
df['HDTIscorenorm'] = df['HDTIscorenorm'].round(2)
df['HDTIscorenorm'].describe()
```

```python
df.sample(10)
```

```python
df['RTI'] = df['Median_Rent'] * 12 / df['Median_Income'] * 100
df['RTIscore'] = 30 - df['RTI']
```

```python
df['RTIscore'].describe()
```

```python
df['RTIscorenorm'] = st.norm.cdf((df['RTIscore'] - df['RTIscore'].mean())/df['RTIscore'].std()) * 100
df['RTIscorenorm'] = df['RTIscorenorm'].round(2)
df['RTIscorenorm'].describe()
```

```python
df.sample(10)
```

```python
df.sample(10)
```

```python
df['MedAgescore'] = st.norm.cdf((df['Median_Age'].mean() - df['Median_Age'])/df['Median_Age'].std()) * 100
df['MedAgescore'] = df['MedAgescore'].round(1)
df['MedAgescore'].describe()
```

```python
df['Povertyscore'] = st.norm.cdf((df['Percent_below_Poverty'].mean() - df['Percent_below_Poverty'])/df['Percent_below_Poverty'].std()) * 100
df['Povertyscore'] = df['Povertyscore'].round(1)
df['Povertyscore'].describe()
```

```python
df['COLIscore'] = st.norm.cdf((df['Cost_of_Living_Index'].mean() - df['Cost_of_Living_Index'])/df['Cost_of_Living_Index'].std()) * 100
df['COLIscore'] = df['COLIscore'].round(1)
df['COLIscore'].describe()
```

```python
df['population_change'].describe()
```

```python
df['popchangescore'] = st.norm.cdf((df['population_change'] - df['population_change'].mean())/df['population_change'].std()) * 100
df['popchangescore'] = df['popchangescore'].round(1)
df['popchangescore'].describe()
```

```python
df.sample(6)
```

```python
df['MedIncscore'] = st.norm.cdf((df['Median_Income'] - df['Median_Income'].mean())/df['Median_Income'].std()) * 100
df['MedIncscore'] = df['MedIncscore'].round(1)
df['MedIncscore'].describe()
```

```python
df['PerCapIncscore'] = st.norm.cdf((df['per_capita_Income'] - df['per_capita_Income'].mean())/df['per_capita_Income'].std()) * 100
df['PerCapIncscore'] = df['PerCapIncscore'].round()
df['PerCapIncscore'].describe()
```

```python
df['MedAgescore'] = st.norm.cdf((df['Median_Age'].mean() - df['Median_Age'])/df['Median_Age'].std()) * 100
df['MedAgescore'] = df['MedAgescore'].round(1)
df['MedAgescore'].describe()
```

```python
df.sample(6)
```

```python
df['PopulationScore'] = (df['popchangescore'] + df['popden_norm'] + df['MedAgescore'])/3
df['EconomyScore'] = (df['MedIncscore'] + df['PerCapIncscore'] + df['Povertyscore'])/3
df['ClimateScore'] = (df['AQIscore'])
df['CostofLivingScore'] = (df['COLIscore'] + df['HDTIscorenorm'] + df['RTIscorenorm'])/3

df['LivabilityScore'] = (df['PopulationScore'] + df['EconomyScore'] + df['ClimateScore'] + df['CostofLivingScore'])/4

df['LivabilityScore'].describe()
```
