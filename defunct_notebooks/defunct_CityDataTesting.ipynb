```python
df1['city'] = df1['city'].str.replace('https://www.city-data.com/city/', '')
df1['city'] = df1['city'].str.replace('.html', '')
df1['city'] = df1['city'].str.replace('Washington-District-of-Columbia', 'Washington, DC')
df1['city'] = df1['city'].str.replace('New-Hampshire', 'New Hampshire')
df1['city'] = df1['city'].str.replace('New-Jersey', 'New Jersey')
df1['city'] = df1['city'].str.replace('New-York', 'New York')
df1['city'] = df1['city'].str.replace('New-Mexico', 'New Mexico')
df1['city'] = df1['city'].str.replace('North-Carolina', 'North Carolina')
df1['city'] = df1['city'].str.replace('North-Dakota', 'North Dakota')
df1['city'] = df1['city'].str.replace('Rhode-Island', 'Rhode Island')
df1['city'] = df1['city'].str.replace('South-Carolina', 'South Carolina')
df1['city'] = df1['city'].str.replace('South-Dakota', 'South Dakota')
df1['city'] = df1['city'].str.replace('West-Virginia', 'West Virginia')

```

```python
result = []
for i in df1['city']:
    source = i
    old = '-'
    new = ', '
    maxreplace = 1
    result.append(new.join(source.rsplit(old, maxreplace)))
    
df1['city'] = result
df1['city'] = df1['city'].str.replace('-', ' ')
df1['city']

```

```python
df1['population_change'] = df1['population_change'].str.replace("%","")
df1['population_change'] = df1['population_change'].astype(float).round(1)
df1['population_change']
```

```python
from sklearn.impute import KNNImputer
```

```python
cols = ['population', 'Population_Density', 'Latitude', 'Longitude', 'Cost_of_Living_Index','Median_Age', 
        'Median_Income', 'per_capita_Income', 'Median_House_Value', 'Unemployment_rate', 'Property_taxes', 
        'Average_Commute_Time']
```

```python
dfdensity = df1[cols]
```

```python
imputer = KNNImputer(n_neighbors=3)
imputedarray = imputer.fit_transform(dfdensity)
```

```python
imputedarray
```

```python
dfimputed = pd.DataFrame(imputedarray, columns=[cols])
dfimputed.head()
```

```python
df1[cols] = dfimputed[cols]
df1.head()
```

```python
cols = cols + ['Median Rent']
```

```python
dfmedrent = df1[cols]

imputedarray = imputer.fit_transform(dfmedrent)

dfimputed = pd.DataFrame(imputedarray, columns=[cols])

df1[cols] = dfimputed[cols]
df1.head()
```

```python
cols = cols + ['AQI']

dfaqi = df1[cols]

imputedarray = imputer.fit_transform(dfaqi)

dfimputed = pd.DataFrame(imputedarray, columns=[cols])

df1[cols] = dfimputed[cols]
df1.head()
```

```python
cols = cols + ['population_change']

dfpopchange = df1[cols]

imputedarray = imputer.fit_transform(dfpopchange)

dfimputed = pd.DataFrame(imputedarray, columns=[cols])

df1[cols] = dfimputed[cols]
df1.head()
```

```python
cols = cols + ['Percent_below_Poverty']

dfpoverty = df1[cols]

imputedarray = imputer.fit_transform(dfpoverty)

dfimputed = pd.DataFrame(imputedarray, columns=[cols])

df1[cols] = dfimputed[cols]
df1.head()
```
