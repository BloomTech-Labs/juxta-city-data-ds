```python
import pandas as pd
import numpy as np
```

```python
filepath = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/07-19-2020.csv'

df = pd.read_csv(filepath)

df.describe()
```

```python
df.head()
```

```python
df.FIPS.isna().sum()
```

```python
df['FIPS'] = df['FIPS'].replace(np.NaN, -1)
```

```python
covid = df[(df['FIPS'] >= 0)]
```

```python
covid1 = covid.drop(columns=['Country_Region', 'Combined_Key'])
```

```python
covid1
```

```python
def get_date(array):
    return array[:10]

get_date(covid1['Last_Update'][0])
```

```python
covid1['Date'] = covid1['Last_Update'].apply(get_date)
covid1.drop(columns=['Last_Update'], inplace=True)
```

```python
covid1.isna().sum()
```

```python
covid1['Active'] = covid1['Active'].fillna(-1)
```

```python
covid1['Active'] = covid1['Active'].astype(int)
```

```python
covid2 = covid1.copy()
```

```python
covid2 = covid2[~covid2['Lat'].isna()==True]
```

```python
covid2['Case-Fatality_Ratio'] = covid2['Case-Fatality_Ratio'].fillna(-1)
```

```python
covid2.isna().sum()
```

```python
len(covid2)
```

```python
covid2['FIPS'] = covid2['FIPS'].astype(int)
```

```python
covid2 = covid2.rename(columns={'Admin2': 'County'})
```

```python
covid2.to_csv('..\\src\\datasets\\covid_data.csv', index=False)
```
