```python
# Load in created CSV

juxta_3 = pd.read_csv('./extraneous_datasets/JuxtaCityData3.csv')
```

```python
# Using a test id to obtain that row in dict format

test_id = 4489
city = juxta_3[juxta_3['id'] == test_id]
output = city.to_dict('records')
print(output)
```

```python
# Obtain value for 'city'

print(output[0]['city'])
```

```python
# Obtain value for 'population'

data = output[0]
population = data['population']
print(f'The population is {population} as of 2017,')
```

```python
# Identifying observations where the 'city' name starts with a given string

test_input = 'Atl'
cities = juxta_3[juxta_3['city'].str.startswith(test_input)]
cities
```

```python
# Instantiate an empty list and fill with the subset's indices

indices = []
for index, row in cities.iterrows():
    indices.append(index)
print(indices)
```
