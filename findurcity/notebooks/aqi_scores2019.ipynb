```python
names = []

for col in air_quality.columns:
    names.append(' '.join(col.split()))

col_names = {}
for old_col, new_col in zip(air_quality.columns, names):
    col_names.update({old_col: new_col})
print(col_names)
```

```python
aq = air_quality.rename(col_names, axis='columns')
```

```python
new_aq = aq.replace({'ND': 0, 'IN': 0})
```

```python
new_aq.dtypes
```

```python
new_aq.head()
```

```python
# Convert to correct units (multiplier dependent on Molecular Weight (M) of molecule)
new_aq['NO2 1-hr'] = new_aq['NO2 1-hr (ppb)'].apply(lambda x: x * 1.88)  # ug/m3
new_aq['SO2 1-hr'] = new_aq['SO2 1-hr (ppb)'].apply(lambda x: x * 2.62)  # ug/m3
new_aq['CO 8-hr'] = new_aq['CO 8-hr (ppm)'].apply(lambda x: 0.0409 * x * 28.01)  # mg/m3 (M = 28.01)
new_aq['O3 8-hr'] = new_aq['O3 8-hr (ppm)'].apply(lambda x: (0.0409 * x * 48) * 1000)  # ug/m3 (M = 48)
new_aq['24-hr PM10'] = new_aq['PM10 24-hr (µg/m3)']  # PM = Particulate Matter

final_aq = new_aq[['State', 'County', 'County FIPS Code', '2010 Population',
                   'CO 8-hr', 'O3 8-hr', 'NO2 1-hr', 'SO2 1-hr', '24-hr PM10'
                  ]]
```

```python
final_aq.describe(exclude='object')
```

```python
co_psi = []
for ob in final_aq['CO 8-hr']:
    if ob <= 5:
        co_psi.append('Good')
    elif (ob > 5) and (ob <= 10):
        co_psi.append('Moderate')
    elif (ob > 10) and (ob <= 17):
        co_psi.append('Unhealthy')
    elif (ob > 17) and (ob <= 34):
        co_psi.append('Very Unhealthy')
    else:
        co_psi.append('Hazardous')
set(co_psi)
```

```python
o3_psi = []
for ob in final_aq['O3 8-hr']:
    if ob <= 119:
        o3_psi.append('Good')
    elif (ob > 119) and (ob <= 157):
        o3_psi.append('Moderate')
    elif (ob > 157) and (ob <= 235):
        o3_psi.append('Unhealthy')
    elif (ob > 235) and (ob <= 785):
        o3_psi.append('Very Unhealthy')
    else:
        o3_psi.append('Hazardous')
set(o3_psi)
```

```python
no2_psi = []
for ob in final_aq['NO2 1-hr']:
    if ob < 1130:
        no2_psi.append('Good')
#     elif (ob > 119) and (ob <= 157):
#         no2_psi.append('Moderate')
    elif (ob > 1130) and (ob <= 2260):
        no2_psi.append('Unhealthy')
    elif (ob > 2260) and (ob <= 2260):
        no2_psi.append('Very Unhealthy')
    else:
        no2_psi.append('Hazardous')
set(no2_psi)
```

```python
so2_psi = []
for ob in final_aq['SO2 1-hr']:
    if ob <= 80:
        so2_psi.append('Good')
    elif (ob > 80) and (ob <= 365):
        so2_psi.append('Moderate')
    elif (ob > 365) and (ob <= 800):
        so2_psi.append('Unhealthy')
    elif (ob > 800) and (ob <= 1600):
        so2_psi.append('Very Unhealthy')
    else:
        so2_psi.append('Hazardous')
set(so2_psi)
```

```python
pm10_psi = []
for ob in final_aq['24-hr PM10']:
    if ob <= 50:
        pm10_psi.append('Good')
    elif (ob > 50) and (ob <= 150):
        pm10_psi.append('Moderate')
    elif (ob > 150) and (ob <= 350):
        pm10_psi.append('Unhealthy')
    elif (ob > 350) and (ob <= 240):
        pm10_psi.append('Very Unhealthy')
    else:
        pm10_psi.append('Hazardous')
set(pm10_psi)
```

```python
final_aq['CO_category'] = co_psi
final_aq['O3_category'] = o3_psi
final_aq['NO2_category'] = no2_psi
final_aq['SO2_category'] = so2_psi
final_aq['PM10_category'] = pm10_psi

final_aq.head()
```
