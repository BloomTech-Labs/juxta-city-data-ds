```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
df['population'].describe()

# Tiny: 5-10k
# Small: 10-30k
# Medium: 30k-100k
# Large: 100k-1 million
# Extra Large: 1 million+
```

```python
def pop_rank(row):
    if row['population'] < 10000:
        val = 1
    elif row['population'] < 30000:
        val = 2
    elif row['population'] < 100000:
        val = 3
    elif row['population'] < 1000000:
        val = 4
    else:
        val = 5
    return val

df2['pop_rank'] = df.apply(pop_rank, axis=1)
```

```python
df['population_change'].describe()

# Decreasing Rapidly < -25
# Decreasing < -5
# Stable < 5
# Increasing >= 5
# Increasing Rapidly >= 25
```

```python
def pop_ch_rank(row):
    if row['population_change'] < -25:
        val = 1
    elif row['population_change'] < -5:
        val = 2
    elif row['population_change'] < 5:
        val = 3
    elif row['population_change'] >= 5:
        val = 4
    else:
        val = 5
    return val

df2['pop_ch_rank'] = df.apply(pop_ch_rank, axis=1)
```
