```
def matchRange(bins):
  if bins in range(0,94):
    return "0-94"
  elif bins in range(95,100):
    return "94.1-100"
  elif bins in range(101,120):
    return "100.1-120"
  elif bins in range(121,140):
    return "120.1-140"
  elif bins in range(141,160):
    return "140.1-160"
  elif bins in range(160,180):
    return "160.1-180"
  elif bins in range(181,200):
    return "181.1-200"
  elif bins in range(200,220):
    return "200.1-220"
  elif bins in range(220,240):
    return "220.1-240"
  elif bins in range(240,260):
    return "240.1-260"
  elif bins in range(260,280):
    return "260.1-280"
  elif bins in range(280,290):
    return "280.1-290"
  elif bins in range(290,300):
    return "290.1-300"
  elif bins in range(310,330):
    return "310.1-330"
  else:
    return ">330.1"
```
